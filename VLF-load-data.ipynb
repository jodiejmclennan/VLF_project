```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load in VLF data .txt file 
reference_file = 'VLFdatatest2.txt'
database = np.loadtxt(reference_file, dtype=float, comments='%') #delimiter='\t'

time = database[:,0] # in seconds
amplitude = database[:,1]
phase = database[:,2]
```

```python
# plot the data
plt.figure(figsize=(8,6),dpi=80)
plt.plot(time, amplitude)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Fairbanks to DHO 2022-03-05')
```
